# 2 Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
president = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
president
```

```python
soup = BeautifulSoup(president.text,'html.parser')
soup
```

```python
president_name = []
for i in soup.find_all('div',class_='presidentListing'):
    president_name.append(i.find("h3").text.split("(")[0])
president_name
```

```python
TOF = []
for i in soup.find_all('div',class_='presidentListing'):
    TOF.append(i.find("p").text.split[0])
TOF
```

```python
print(len(president_name),len(TOF))
```

```python
import pandas as pd
df_pr=pd.DataFrame({'President Name':president_name,'TOF':TOF})
df_pr
```

Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame

i) Headline
ii) Time
iii) News Link

```python
news = requests.get('https://www.cnbc.com/world/?region=world')
news
```

```python
soup = BeautifulSoup(news.text,'html.parser')
soup
```

```python
headline = []
for i in soup.find_all('div',class_='LatestNews-container'):
    headline.append(i.find('a').text.split("(")[-1])
headline
```

```python
time = []
for i in soup.find_all('span',class_='LatestNews-wrapper'):
    time.append(i.text)
time
```

```python
link = []
for i in soup.find_all('div',class_='LatestNews-headline'):
    link.append(i.get('href'))
link
```

```python
print(len('headline'),len('time'))
```

```python
df_news=pd.DataFrame({'headline':headline,'time':time})
df_news
```

Write a python program to scrape mentioned details from dineout.co.in and make data frame

i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
restaurant = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
restaurant
```

```python
soup=BeautifulSoup(restaurant.content)
soup
```

```python
name = []

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)
    
name
```

```python
Cuisine = []

for i in soup.find_all('div',class_="main-content form-elements-wrap filter-options-wrap"):
    Cuisine.append(i.text)
Cuisine
```

```python
Location = []
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    Location.append(i.text)
Location
```

```python
Ratings = []

for i in soup.find_all('div',class_='restnt-rating rating-4'):
    Ratings.append(i.text)
Ratings
```

```python
images = []
for i in soup.find_all('img',class_="no-img"):
    images.append(i.get('data-src'))
images
```

```python
df_res = pd.DataFrame({'name':name,'Cuisine':Cuisine,'Location':Location,'Ratings':Ratings,'images':images})
df_res
```
